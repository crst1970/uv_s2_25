```python
# Importar las librerias a utilizar
import numpy as np
import matplotlib.pyplot as plt
import pydicom
import statsmodels
from os import listdir
```

```python
# Lectura de imagen 
#data_img_1 = pydicom.dcmread('img_dcm_1.dcm')
data_img_1 = pydicom.dcmread('IM-0001-0001-0001.dcm') 
img_1 = data_img_1.pixel_array
img_1_head = data_img_1 #acceso a los metadatos en el header

print('Tamano de imagen: ',img_1.shape)
print('Data type de la imagen: ',img_1.dtype)


#visualization
plt.figure()
plt.imshow(img_1, cmap=plt.cm.gray)  
plt.xlabel('tamaño (px)')
plt.ylabel('tamaño (px)')
plt.title('Imagen 1')
plt.colorbar() #IMPORTANTE : confirmar los valores visualizados
plt.show()
```

```python
#Cambio de la escala de visualizacion

f = plt.figure()
f = plt.imshow(img_1, cmap=plt.cm.gray)
f.set_clim(0,200)   #setea los limites de visualizacion: por debajo y por arriba de estos valores "satura" la visualizacion
plt.title('Imagen 1 visualizada de otra manera')
plt.colorbar() #IMPORTANTE : confirmar los valores visualizados
plt.show()
```
