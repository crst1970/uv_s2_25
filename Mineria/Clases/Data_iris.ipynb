```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',header= None)  # Read the Iris dataset from a CSV file
data.columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'class']  # Assign column names to the DataFrame
data.head()  # Display the first five rows of the DataFrame
```

```python
import matplotlib.pyplot as plt

for i, column in enumerate(data.columns[:-1]):  # Excluir la última columna (class)
    plt.figure(figsize=(8, 6))
    data[column].hist(bins=10)
    plt.title(f'Histograma de {column}')
    plt.xlabel(column)
    plt.ylabel('Frecuencia')
    plt.show()
```

```python
data.boxplot()
```

```python
for i, column in enumerate(data.columns[:-1]):  
    plt.figure(figsize=(8, 6))
    plt.scatter(data.index, data[column].values, label=column)
    plt.title(f'Diagrama de dispersión de {column}')
    plt.xlabel('Índice')
    plt.grid()
    plt.ylabel(column)
    plt.legend()
    plt.show()
```

```python
import seaborn as sns
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
k = 0

for i in range(2):
    for j in range(2):
        sns.histplot(x = data.columns[k], data=data, ax=axes[i, j], kde=True, hue='class', bins=10)
        axes[i, j].set_title(f'Histograma de {data.columns[k]}')
        axes[i, j].set_xlabel(data.columns[k])
        axes[i, j].set_ylabel('Frecuencia')
        axes[i, j].grid()
        k += 1
plt.tight_layout()  
plt.show()
```

```python
sns.FacetGrid(data, hue="class", height=5).map(sns.scatterplot, "sepal length", "sepal width").add_legend()
plt.title('Diagrama de dispersión de Sepal Length vs Sepal Width')
plt.grid()
plt.show()
```
